```python
# import
import numpy as np
import matplotlib.pyplot as plt
import os
import json
import scienceplots
```

```python
# create class for data srtucture
class dataStructure:
    # init function, for angles, closest_angles, delta, infofile
    def __init__(self, name, numberOfRotations, numberOfLaser, absLaserMovement):
        self.name = name[:-11]
        self.numberOfRotations = numberOfRotations
        self.numberOfLaser = numberOfLaser
        self.absLaserMovement = absLaserMovement
        
```

```python
def readData(folderName):
    # import all files from the LogFiles folder
    path = 'LogFiles/' + folderName + '/'
    files = os.listdir(path)

    # remove all files not ending with .json
    files = [file for file in files if file.endswith('.json')]

    # sort files by name
    files.sort()


    print(files)
    # create a dictionary for the data
    data = {}

    # loop over all files
    for file in files:
        # open file
        with open(path + file) as json_file:
            # read data from file
            data_file = json.load(json_file)
            
            # create data structure
            data[file] = dataStructure(file, data_file['numberOfRotations'], data_file['numberOfLaser'], data_file['absLaserMovement'])
        

    # print all names
    for key in data:
        print(data[key].name)

    return data
```

```python
# set matplotlib style

# set the style for matplotlib to scientific
plt.style.use('science')
# set size of the figure
plt.rcParams["figure.figsize"] = (12, 6)
# set the font size
plt.rcParams.update({'font.size': 15})
```

```python
def visualizeData(data, folderName):
    # create a plot for every name, and add bar subplots for every numberOfLasers which includes the NumberOfRotations and the absLaserMovement
    # there shall be a subplot for numberOfRotations and a subplot for absLaserMovement, the X-Axis shall be the numberOfLasers

    # create figure
    fig, axs = plt.subplots(1, 2)

    # set title for the whole figure
    fig.suptitle(folderName)

    # add vertical space between subplots
    fig.subplots_adjust(hspace=0.5)

    ### subplot number of rotation
    # array for numberOfRotations
    numberOfRotations = []
    # array for absLaserMovement
    absLaserMovement = []
    # array for numberOfLaser
    numberOfLaser = []
    # loop over all files
    for key in data:
        # append to array but convert first to float than to int
        numberOfRotations.append(int(float(data[key].numberOfRotations)))
        # append to array but convert first to float than to int
        absLaserMovement.append(int(float(data[key].absLaserMovement)))
        # append to array but convert first to float than to int
        numberOfLaser.append(int(float(data[key].numberOfLaser)))


    # create bar plot for numberOfRotations, numberOfLaser is a string and needs to be displayed as a string
    Labels = [str(number) for number in numberOfLaser]
    axs[0].bar(Labels, numberOfRotations)


    # set title
    axs[0].set_title('numberOfRotations')
    # set x label
    axs[0].set_xlabel('numberOfLaser')
    # set y label
    axs[0].set_ylabel('numberOfRotations (-)')
    # start y axis at 0
    axs[0].set_ylim(bottom=0)

    # add grid both
    axs[0].grid(axis='both')

    # # add horizontal space between subplots
    fig.subplots_adjust(wspace=0.5)
    

    ### subplot absLaserMovement
    axs[1].bar(Labels, absLaserMovement)

    # set title
    axs[1].set_title('absLaserMovement')
    # set x label
    axs[1].set_xlabel('numberOfLaser')
    # set y label
    axs[1].set_ylabel('absLaserMovement (kÂ°)')   
    # start y axis at 0
    axs[1].set_ylim(bottom=0)


    # add grid both
    axs[0].grid(axis='both')


    # save figure
    plt.savefig('LogFiles/' + folderName + '/' + folderName + '.png', dpi=300, bbox_inches='tight')
```

```python
# names of the folder
folderNames = ['Cuboid', 'Cylinder', 'Hexagon', 'Freeform']

# loop over all folder names
for folderName in folderNames:
    # read data from folder
    data = readData(folderName)
    # visualize data
    visualizeData(data, folderName)
    # save figure
    plt.savefig(folderName + '.png')
    # show figure
    plt.show()
```
